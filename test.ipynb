```python
import numpy as np
import matplotlib.pyplot as plt

# Increase the upper bound from 20 to 100
n = np.linspace(2, 100, 1000)

O1      = np.ones_like(n)        
O_log_n = np.log(n)              
O_n     = n                      
O_nlogn = n * np.log(n)          
O_n2    = n**2                   
O_2n    = 2**n                   

plt.figure(figsize=(12, 8))

# Fill regions
plt.fill_between(n, O_log_n, O_n,     color='green', alpha=0.3, label=r'Region: $O(n)$')
plt.fill_between(n, O_n,     O_nlogn, color='yellow', alpha=0.3, label=r'Region: $O(n \log n)$')
plt.fill_between(n, O_nlogn, O_n2,    color='orange', alpha=0.3, label=r'Region: $O(n^2)$')
plt.fill_between(n, O_n2,    O_2n,    color='red', alpha=0.3,    label=r'Region: $O(2^n)$')
plt.fill_between(n, O1,      O_log_n, color='lightblue', alpha=0.3, label=r'Region: $O(\log n)$')

# Plot boundaries
plt.plot(n, O1,      'k--', label=r'$O(1)$')
plt.plot(n, O_log_n, 'k--', label=r'$O(\log n)$')
plt.plot(n, O_n,     'k--', label=r'$O(n)$')
plt.plot(n, O_nlogn, 'k--', label=r'$O(n \log n)$')
plt.plot(n, O_n2,    'k--', label=r'$O(n^2)$')
plt.plot(n, O_2n,    'k--', label=r'$O(2^n)$')

plt.xlabel('Input size (n)')
plt.ylabel('Operations')
plt.title('Big-O Complexity Regions')
plt.xscale('log')  
plt.yscale('log')  
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend(loc='upper left')

plt.show()

```
